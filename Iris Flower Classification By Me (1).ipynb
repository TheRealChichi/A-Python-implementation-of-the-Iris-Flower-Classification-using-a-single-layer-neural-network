```python
import random
random.seed(42)
test_setosa = random.sample(list(setosa.index), int(len(setosa) * 0.1))
test_versicolor = random.sample(list(versicolor.index), int(len(versicolor) * 0.1))
test_virginica = random.sample(list(virginica.index), int(len(virginica) * 0.1))

test_indices = test_setosa + test_versicolor + test_virginica
```

```python
test_data = df.loc[test_indices]
train_data = df.drop(test_indices)
```

```python
X_train = train_data.iloc[:, :-1].values.astype(float)
y_train = train_data.iloc[:, -1].values
X_test = test_data.iloc[:, :-1].values.astype(float)
y_test = test_data.iloc[:, -1].values
```

```python
class_mapping = {'Iris-setosa':0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
y_train_encoded = np.array([class_mapping[label] for label in y_train])
```

```python
def one_hot_encode(Y, num_classes):
    one_hot = np.zeros((Y.size, num_classes))
    one_hot[np.arange(Y.size), Y] = 1
    return one_hot

y_encoded = one_hot_encode(y_train_encoded, 3)
```

```python
def initialisation(X):
    W = np.random.randn(X.shape[1], 3)
    b = np.random.randn(1,3)
    return(W, b)
```

```python
def softmax(Z):
    expZ = np.exp(Z - np.max(Z))
    return expZ / expZ.sum(axis = 1, keepdims = True)
```

```python
def model(X, W, b):
    Z = X.dot(W) + b 
    A =  softmax(Z)
    return A
```

```python
def cross_entropy_loss(A, Y):
    m = Y.shape[0]
    epsilon = 1e-15
    A = np.clip(A,epsilon, 1 - epsilon)
    log_likelihood = -np.sum(Y * np.log(A))
    return log_likelihood / m
```

```python
def gradient(A,X, y):
    m = y.shape[0]
    dW = (1/m) * np.dot(X.T, A - y)
    db = (1/m) * np.sum(A - y, axis =  0, keepdims = True)
    return dW, db
```

```python
def update(dW, db, W, b, learning_rate):
    W = W - learning_rate * dW
    b = b - learning_rate * db
    return (W,b)
```

```python
def artificial_neurone(X_train, Y_train, learning_rate=0.1, n_iter=1000):
    # Initialize weights and bias
    W, b = initialisation(X_train)
    
    loss = []

    for i in range(n_iter):
        # Forward propagation
        A = model(X_train, W, b)

        # Compute loss
        loss_value = cross_entropy_loss(A, Y_train)
        loss.append(loss_value)

        # Backward propagation
        dW, db = gradient(A, X_train, Y_train)

        # Update weights
        W -= learning_rate * dW
        b -= learning_rate * db

    # Plot loss over time
    import matplotlib.pyplot as plt
    plt.plot(loss)
    plt.xlabel("Iterations")
    plt.ylabel("Cross-Entropy Loss")
    plt.title("Training Loss")
    plt.show()

    return W, b

# Train the model
W, b = artificial_neurone(X_train, y_encoded)
```

```python
def predict(X, W, b):
    Z = np.dot(X, W) + b
    A = softmax(Z)
    predictions = np.argmax(A, axis=1)
    
    return predictions
```

```python
predictions = predict(X_test, W, b)
print("Predicted classes:", predictions)
```
